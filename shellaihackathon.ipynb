```
import pandas as pd
import numpy as np
 
df = pd.read_csv('MOCK_DATA_SHELL.csv')
df.head()
```

**TRANSFORM THE DATA TO MAKE IT USABLE FOR LOGISTIC REGRESSION**

```
df2 =df
df2.head()
```

**CONVERTING ALL VAUES TO RISK SCORES**

```
df2['Primary contact'] = df2['Primary contact'].replace([True,False],[7,0])
df2.head()
```

```
df2.describe()
```

```
df2['Covid Awarness quiz score'] = np.where(
   (df2['Covid Awarness quiz score'] <=7)  , 7, df2['Covid Awarness quiz score']
   )

df2['Covid Awarness quiz score'] = np.where(
   (df2['Covid Awarness quiz score'] <=11) & (df2['Covid Awarness quiz score'] > 7) , 4, df2['Covid Awarness quiz score']
   )

df2['Covid Awarness quiz score'] = np.where(
   (df2['Covid Awarness quiz score'] >11)  , 1, df2['Covid Awarness quiz score']
   )

df2['Covid Awarness quiz score']
```

```
df2['Population density'] = np.where(
   (df2['Population density'] <=60000)  , 1, df2['Population density']
   )

df2['Population density'] = np.where(
   (df2['Population density'] <=120000) & (df2['Population density'] > 60000) , 3, df2['Population density']
   )

df2['Population density'] = np.where(
   (df2['Population density'] >120000)  , 5, df2['Population density']
   )

df2['Population density']
```

```
df2['Number of times left home'] = df2['Number of times left home'] / 3 
df2['Number of times left home'] = np.where(
   (df2['Number of times left home'] <=7)  , 1, df2['Number of times left home']
   )

df2['Number of times left home'] = np.where(
   (df2['Number of times left home'] <=21) & (df2['Number of times left home'] > 7) , 3, df2['Number of times left home']
   )

df2['Number of times left home'] = np.where(
   (df2['Number of times left home'] >21)  , 5, df2['Number of times left home']
   )

df2['Number of times left home']
```

```
df2['Tested Positive and recovered'] = df2['Tested Positive and recovered'].replace([True,False],[1,0]) #the output column to be modelled
df2['Secondary contact'] = df2['Secondary contact'].replace([True,False],[5,0])
df2['Have you left India in the last 2 months'] = df2['Have you left India in the last 2 months'].replace([True,False],[3,0])
df2['Previously infected with COVID'] = df2['Previously infected with COVID'].replace([1,2,3,4],[1,3,5,0])
df2['Priority to available at work place'] = df2['Priority to available at work place'].replace([1,2,3,4],[1,3,5,7])
df2.head()
```

```
df2.head()
```

```
df2['Number of people at home'] = df2['Number of people at home'] /10
df2['Number of people at home'] = np.where(
   (df2['Number of people at home'] <=2)  , 1, df2['Number of people at home']
   )

df2['Number of people at home'] = np.where(
   (df2['Number of people at home'] <=5) & (df2['Number of people at home'] > 2) , 3, df2['Number of people at home']
   )

df2['Number of people at home'] = np.where(
   (df2['Number of people at home'] >5)  , 5, df2['Number of people at home']
   )

df2['Number of people at home']
```

```
df2['Total frequency of leaving home (Family)'] = df2['Total frequency of leaving home (Family)'] /2
df2['Total frequency of leaving home (Family)'] = np.where(
   (df2['Total frequency of leaving home (Family)'] <=7)  , 1, df2['Total frequency of leaving home (Family)']
   )

df2['Total frequency of leaving home (Family)'] = np.where(
   (df2['Total frequency of leaving home (Family)'] <=21) & (df2['Total frequency of leaving home (Family)'] > 7) , 3, df2['Total frequency of leaving home (Family)']
   )

df2['Total frequency of leaving home (Family)'] = np.where(
   (df2['Total frequency of leaving home (Family)'] >21)  , 5, df2['Total frequency of leaving home (Family)']
   )

df2['Total frequency of leaving home (Family)']

#df2.head()
```

**CREATE DICTIONARY OF IMPORTANCE OF EACH FACTOR AND CALCULATE THE SCORES**

```
risk_factors = {'Primary contact' : 10, 'Previously infected with COVID' : 10,	'Covid Awarness quiz score' : 5,	
                'Population density' : 5,	'Number of times left home' : 5, 
                 'Secondary contact' : 5,	'Priority to available at work place' : 5,
                 'Existing health condition' : 10,	'Number of people at home' : 5,
                'Have you left India in the last 2 months' : 5,	'Total frequency of leaving home (Family)' : 1}
risk_factors
```

```
data = df2.copy()
```

```
data
```

```
scores = []
for index, row in data.iterrows():
    score = 0
    for key in risk_factors.keys(): 
      score += row[key] * risk_factors[key]
      print(row[key],score, end="  ")
    print("\n\n\n\n\n\n")
    scores.append(score)  
```

```

for key in risk_factors.keys(): 
  data[key] = data[key] * risk_factors[key]

data
```

```
data
```

```
print("Average score - " + str(sum(scores)/len(scores)))
print("Min score - " + str(min(scores)))
print("Max score - " + str(max(scores)))
```

```
df_scores = pd.Series(scores)
df_scores.describe()
```

```
df_scores.to_csv("scores.csv")
```
